```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#avg_tuition_file
avg_tuition_file = "static/data/avg_tuiton_state.csv"
avg_tuition_df = pd.read_csv(avg_tuition_file)
avg_tuition_df.head()
```

```python
#income_by_education_level
income_lvleducation_file = "static/data/income_by_education_level.csv"
income_lvleducation_df = pd.read_csv(income_lvleducation_file)
income_lvleducation_df.head()
```

```python
#income_by_state
state_income_file = "static/data/income_by_state.csv"
state_income_df = pd.read_csv(state_income_file)
state_income_df.head()
```

```python
#Kaggle_tuition_cost
tuition_cost_file = "static/data/Kaggle_tuition_cost.csv"
tuition_cost_df = pd.read_csv(tuition_cost_file)
tuition_cost_df.head()
```

```python
#State_Avg_Debt
state_debt_file = "static/data/State-Avg-Debt.csv"
state_debt_df = pd.read_csv(state_debt_file)
state_debt_df.head()
```

```python
#State_Med_Income
State_Med_Income_file = "static/data/State-Med-Income-Edu.csv"
State_Med_Income_df = pd.read_csv(State_Med_Income_file)
State_Med_Income_df.head()
```

```python
#unemployment_by_state
unemployment_state_file = "static/data/unemployment_by_state.csv"
unemployment_state_df = pd.read_csv(unemployment_state_file)
unemployment_state_df.head()
```

```python
connection_string = "postgres:postgres@localhost:5432/Proj3-Team3_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#avg_tuition
avg_tuition_df.to_sql(name='premise', con=engine, if_exists='append', index=True)
#income_by_education_level
income_lvleducation_df.to_sql(name='premise', con=engine, if_exists='append', index=True)
#income_by_state
state_income_df.to_sql(name='premise', con=engine, if_exists='append', index=True)
#Kaggle_tuition_cost
tuition_cost_df.to_sql(name='premise', con=engine, if_exists='append', index=True)
#State_Avg_Debt
state_debt_df.to_sql(name='premise', con=engine, if_exists='append', index=True)
#State_Med_Income
State_Med_Income_df.to_sql(name='premise', con=engine, if_exists='append', index=True)
#unemployment_by_state
unemployment_state_df.to_sql(name='premise', con=engine, if_exists='append', index=True)
```
